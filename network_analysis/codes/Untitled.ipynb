```python
import pandas as pd
import numpy as np
import json 
import math
import collections
import networkx as nx
from matplotlib import pyplot as plt
```

```python
G = nx.read_weighted_edgelist('../data/years/2020.weighted.edgelist')
```

```python
len(G.nodes)
```

```python
weights = sorted([G.edges[edge]['weight'] for edge in G.edges()], reverse=True)
```

```python
for edge in G.edges():
    G.edges[edge]['weight'] = 1 / G.edges[edge]['weight'] 
```

```python
weightsCount = collections.Counter(weights)
w, cnt = zip(*weightsCount.items())
```

```python
plt.plot(w,cnt)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('weight')
plt.ylabel('number of edges')
```

```python
cumcount = [0]
for i, c in enumerate(cnt):
    cumcount.append(c + cumcount[-1])
cumcount = cumcount[1:]
plt.plot(w, cumcount, linestyle = '-')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('weight')
plt.ylabel('cumulative number of edges')
```
